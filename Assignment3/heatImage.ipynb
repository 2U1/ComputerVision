# HeatImage

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import time
import numpy as np
```

```python
image = plt.imread('image3.png')
# Converting the image to grayscale
R, G, B = image[:,:,0], image[:,:,1], image[:,:,2]
image = 0.2989 * R + 0.5870 * G + 0.1140 * B
plt.imshow(image, cmap='gray')
plt.show()
```

```python
def update_by_step(img, i,j):
    # Calculating the vlaue via heat equation from the pdf file
    # delta_x and delta_y is noted 1 from the pdf file
    laplacian = img[i, j-1] + img[i-1,j] + img[i, j+1] + img[i+1, j] - 4*img[i,j]
    return img[i,j] + 0.1 * laplacian
```

```python
def heatEquation(img):
    # pad the image to calculate the values in the edges of the image
    pad_img = np.pad(img, 1, mode='constant')
    cur_image = pad_img.copy()
    # iterate for 100 times
    for L in range(1,101):
        next_img = np.zeros(pad_img.shape)
        for i in range(1, pad_img.shape[0]-1):
            for j in range(1, pad_img.shape[1]-1):
                # update the pixels with the given heat equation
                next_img[i,j] = update_by_step(cur_image, i, j)
        
        # plotting the image every 10 time steps 
        if L % 10 == 0:
            plt.imshow(cur_image, cmap='gray')
            plt.show()

        cur_image = next_img
```

```python
heatEquation(image)
```
